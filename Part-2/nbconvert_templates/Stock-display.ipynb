Given a stock symbol and a number of days, plot its daily price and a 30-day moving average. Code cribbed from [this notebook](http://nbviewer.ipython.org/github/twiecki/financial-analysis-python-tutorial/blob/master/1.%20Pandas%20Basics.ipynb) by [Thomas Wiecki](https://github.com/twiecki).

```python
stock = 'YHOO'   # Display names are stored in notebook metadata
days_back = 600
```

```python
import datetime

import pandas as pd
import pandas.io.data as webdata
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
now = datetime.date.today()
start = now - datetime.timedelta(days=days_back)
df = webdata.get_data_yahoo(stock, start=start, end=now)
```

```python
close_px = df['Adj Close']
mavg = pd.rolling_mean(close_px, 30)
```

```python
close_px.plot(label=stock)
mavg.plot(label='mavg')
plt.legend()
```
